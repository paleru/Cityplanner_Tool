```python
swe_flux_df = pd.read_csv('data/net_emissions/SWE_cities.csv')
swe_flux_df = swe_flux_df.drop(columns=['region_id', 'city_id'])
swe_flux_df['region_name'] = swe_flux_df['region_name'].apply(dh.replace_special_chars)
swe_flux_df['city_name'] = swe_flux_df['city_name'].apply(dh.replace_special_chars)
swe_flux_df
```

```python
fin_reg_emissions_df = pd.read_csv('data/transformed_finland_data/finland_regions_emissions.csv')
fin_reg_emissions_df
```

```python
#adjust regiona names for consistency
fin_flux_df = fin_flux_df.drop(columns=['region_name', 'region_id', 'city_id'])
fin_flux_df['city_name'] = fin_flux_df['city_name'].str.replace('Southern', 'South')
fin_flux_df['city_name'] = fin_flux_df['city_name'].str.replace('Northern', 'North')
fin_flux_df['city_name'] = fin_flux_df['city_name'].str.replace('Savonia', 'Savo')
fin_flux_df = fin_flux_df.replace(['Finland Proper', 'Tavastia Proper', 'Päijänne Tavastia'], ['Southwest Finland', 'Kanta-Haeme', 'Paeijaet-Haeme'])
```

```python
# combine uusimaa and eastern uusimaa emissions
uusimaa_value = fin_flux_df.loc[fin_flux_df['city_name'] == 'Uusimaa', 'net_forest_emissions_01_to_23'].values[0]
eastern_uusimaa_value = fin_flux_df.loc[fin_flux_df['city_name'] == 'Eastern Uusimaa', 'net_forest_emissions_01_to_23'].values[0]

# update uusimaa
fin_flux_df.loc[fin_flux_df['city_name'] == 'Uusimaa', 'net_forest_emissions_01_to_23'] = uusimaa_value + eastern_uusimaa_value

# drop eastern uusimaa
fin_flux_df = fin_flux_df[fin_flux_df['city_name'] != 'Eastern Uusimaa']
```

```python
# assign all the paeijaet-haeme rows to new df and sum their emission values
fin_flux_df_paeijaet = fin_flux_df.loc[fin_flux_df['city_name'] == 'Paeijaet-Haeme']
fin_flux_df_paeijaet['net_forest_emissions_01_to_23'] = fin_flux_df_paeijaet['net_forest_emissions_01_to_23'].sum() 
```

```python
fin_flux_df_paeijaet
```

```python
# replace paeijaet-hame values in original df and drop surplus
fin_flux_df.loc[fin_flux_df['city_name'] == 'Paeijaet-Haeme', 'net_forest_emissions_01_to_23'] = fin_flux_df_paeijaet
fin_flux_df = fin_flux_df.drop_duplicates(subset=['city_name'])
```

```python
# manually add aaland as it was not included in dataset
aaland = {'city_name': 'Aaland', 'net_forest_emissions_01_to_23': -3570000}
fin_flux_df = fin_flux_df._append(aaland, ignore_index=True)
```

```python
# divide by 23 to get yearly average
fin_flux_df['net_forest_emissions_01_to_23'] = fin_flux_df['net_forest_emissions_01_to_23']/23
fin_flux_df = fin_flux_df.rename(columns={'city_name': 'Region', 'net_forest_emissions_01_to_23': 'Average Yearly Net Tree Emissions(For Whole Region)'})

swe_flux_df['net_forest_emissions_01_to_23'] = swe_flux_df['net_forest_emissions_01_to_23']/23
swe_flux_df = swe_flux_df.rename(columns={'region_name': 'Region', 'net_forest_emissions_01_to_23': 'Average Yearly Net Tree Emissions', 'city_name': 'City'})
```

```python
# dict containing regiona name and number of cities and region as k/v pair
reg_num_dict = {
    'Uusimaa': 16,
    'Lapland': 7,
    'Satakunta': 4,
    'Kanta-Haeme': 4,
    'South Ostrobothnia': 3,
    'Southwest Finland': 3,
    'South Savo':3,
    'North Ostrobothnia':3,
    'Paeijaet-Haeme':2,
    'Kymenlaakso': 2,
    'North Savo': 2,
    'Pirkanmaa': 2,
    'South Karelia': 2,
    'Central Finland': 1,
    'Aaland': 1,
    'Central Ostrobothnia': 1,
    'Kainuu': 1,
    'North Karelia': 1,
    'Ostrobothnia': 1,
}

# map using dict
fin_flux_df['Num of Cities in Region'] = fin_flux_df['Region'].map(reg_num_dict)

# do the division based on previously mapped divisor
fin_flux_df['Average Yearly Net Tree Emissions'] = fin_flux_df['Average Yearly Net Tree Emissions(For Whole Region)'] / fin_flux_df['Num of Cities in Region']
fin_flux_df
```
